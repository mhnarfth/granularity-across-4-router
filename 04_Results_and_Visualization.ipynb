```python
#
#  notebooks/04_Results_and_Visualization.ipynb
#

################################################################################
#
# Notebook 4: Results Analysis & Visualization
#
# --- PURPOSE ---
# This is the final notebook in the pipeline. Its purpose is to:
#   1. Load the master metrics CSV and detailed prediction files generated by
#      `03_Experiments.ipynb`.
#   2. Analyze the comprehensive results to answer our core research questions.
#   3. Generate the final, publication-quality figures and tables that will
#      form the results section of the journal manuscript.
#
# --- WORKFLOW ---
# This notebook should be run ONLY AFTER `03_Experiments.ipynb` is complete.
# It is designed to be run cell-by-cell. Each block generates a specific
# figure or table and has its own aesthetic knobs for easy customization.
#
################################################################################


#######################################################################
# CELL 1: ENVIRONMENT SETUP & CONFIGURATION (KNOBS)
#######################################################################

# --- Imports ---
import warnings
from pathlib import Path

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker
from IPython.display import display

print("--- Environment Setup ---")
print(f"Pandas version: {pd.__version__}")

# --- Suppress Warnings ---
warnings.filterwarnings("ignore")

# --- Matplotlib and Seaborn Configuration ---
%matplotlib inline
sns.set_style("whitegrid")
plt.rcParams['figure.dpi'] = 150
plt.rcParams['savefig.dpi'] = 300

# ---------------------------------------------------------------------------------
# >>> CONFIGURATION KNOBS <<<
# ---------------------------------------------------------------------------------

# --- Input Paths ---
METRICS_CSV_PATH = Path("outputs/results/all_granularity_study_results.csv")
PREDICTIONS_DIR = Path("outputs/predictions/")

# --- Output Path ---
FINAL_FIGURES_DIR = Path("outputs/results/figures/")

# --- Analysis Parameters ---
PRIMARY_METRIC = 'sMAPE'
FOCAL_HORIZON_HOURS = 12
HORIZONS_HOURS = [1, 6, 12, 24] # The full list of horizons that were tested

# ---------------------------------------------------------------------------------

# --- Helper Function to Save Plots (PDF ONLY) ---
def save_and_display_plot(fig_name: str):
    """Saves the current matplotlib figure as a PDF and displays it."""
    FINAL_FIGURES_DIR.mkdir(parents=True, exist_ok=True)
    pdf_path = FINAL_FIGURES_DIR / f"{fig_name}.pdf"
    plt.tight_layout()
    plt.savefig(pdf_path)
    print(f"  ✅ Plot saved to: {pdf_path}")
    plt.show()
    plt.close()

print("\n--- Configuration Loaded ---")
print(f"Input Metrics CSV: {METRICS_CSV_PATH}")
print(f"Input Predictions Dir: {PREDICTIONS_DIR}")
print(f"Output Figures Directory: {FINAL_FIGURES_DIR}")
print(f"Primary metric for analysis: '{PRIMARY_METRIC}' at {FOCAL_HORIZON_HOURS}h horizon.")
print("--------------------------------\n")


#######################################################################
# CELL 2: LOAD MASTER METRICS DATA & DEFINE HELPERS
#######################################################################

print(f"--- Loading master metrics from {METRICS_CSV_PATH.name}... ---")

if not METRICS_CSV_PATH.exists():
    raise FileNotFoundError(f"[FATAL ERROR] Master metrics file not found at: {METRICS_CSV_PATH}. Please run 03_Experiments.ipynb first.")

try:
    df_results = pd.read_csv(METRICS_CSV_PATH)
    categorical_cols = ['router', 'granularity', 'model']
    for col in categorical_cols:
        df_results[col] = df_results[col].astype('category')
    print(f"  ✅ Master metrics loaded successfully. Found {len(df_results):,} result entries.")
    
    granularity_order = ['5min', '15min', '30min', '45min', '1H', '2H', '4H', '6H', '12H', '1D']
    df_results['granularity'] = pd.Categorical(df_results['granularity'], categories=granularity_order, ordered=True)
    
except Exception as e:
    print(f"[FATAL ERROR] Could not load or process the metrics file: {e}")
    df_results = pd.DataFrame()

# --- START OF FIX: Define Helper Function here ---
def hours_to_steps(hours: int, granularity: str) -> int:
    """Dynamically calculates the number of steps corresponding to a number of hours."""
    delta_per_step = pd.to_timedelta(granularity)
    return max(1, int(pd.to_timedelta(f'{hours}H') / delta_per_step))
# --- END OF FIX ---

print("--- Helper functions defined. ---")


##########################################################################################
# --- BLOCK FOR FIGURE 1: Granularity's Impact on Prediction Error ---
##########################################################################################

#######################################################################
# CELL 3: Aesthetic Knobs for Figure 1
#######################################################################
FIG1_FIGSIZE = (16, 12)
FIG1_SUPTITLE = f'Impact of Granularity on Forecasting Performance ({FOCAL_HORIZON_HOURS}h Horizon)'
FIG1_XLABEL = 'Time Series Granularity'
FIG1_YLABEL = f'Prediction Error ({PRIMARY_METRIC})'
FIG1_TITLE_FS = 20
FIG1_LABEL_FS = 16
FIG1_TICK_FS = 12
FIG1_LEGEND_FS = 12

#######################################################################
# CELL 4: Code for Figure 1
#######################################################################
if not df_results.empty:
    print("\n--- Generating Figure 1: Granularity vs. Performance ---")
    df_fig1 = df_results[df_results['horizon_hours'] == FOCAL_HORIZON_HOURS].copy()
    fig, axes = plt.subplots(2, 2, figsize=FIG1_FIGSIZE, sharey=True)
    axes = axes.flatten()
    routers = df_fig1['router'].unique().categories
    for i, router in enumerate(routers):
        ax = axes[i]
        df_router = df_fig1[df_fig1['router'] == router]
        sns.lineplot(data=df_router, x='granularity', y=PRIMARY_METRIC, hue='model', marker='o', ax=ax)
        ax.set_title(f'Router: {router.capitalize()}', fontsize=FIG1_LABEL_FS)
        ax.set_xlabel(FIG1_XLABEL, fontsize=FIG1_TICK_FS)
        ax.set_ylabel(FIG1_YLABEL if i % 2 == 0 else '', fontsize=FIG1_TICK_FS)
        ax.tick_params(axis='x', rotation=45)
        ax.grid(True, which='both', linestyle='--', linewidth=0.5)
        ax.legend(title='Model', fontsize=FIG1_LEGEND_FS - 2)
    fig.suptitle(FIG1_SUPTITLE, fontsize=FIG1_TITLE_FS, weight='bold')
    save_and_display_plot("fig_01_granularity_vs_performance")
else:
    print("\n[SKIPPED] Figure 1: Results DataFrame is empty.")


##########################################################################################
# --- BLOCK FOR FIGURE 2: Model Ablation & Baseline Comparison ---
##########################################################################################

#######################################################################
# CELL 5: Aesthetic Knobs for Figure 2
#######################################################################
FIG2_FIGSIZE = (16, 12)
FIG2_SUPTITLE = f'Model Architecture Performance at Optimal Granularity ({FOCAL_HORIZON_HOURS}h Horizon)'
FIG2_XLABEL = 'Model Architecture'
FIG2_YLABEL = 'Prediction Error (MAE)'
FIG2_TITLE_FS = 20
FIG2_LABEL_FS = 16
FIG2_TICK_FS = 12

#######################################################################
# CELL 6: Code for Figure 2
#######################################################################
if not df_results.empty:
    print("\n--- Generating Figure 2: Model Performance at Optimal Granularity ---")
    df_fig2_base = df_results[df_results['horizon_hours'] == FOCAL_HORIZON_HOURS].copy()
    optimal_grans = {}
    df_for_optimals = df_fig2_base[df_fig2_base['model'] == 'hybrid_gru_lstm'].copy()
    for router in df_for_optimals['router'].unique():
        df_router_only = df_for_optimals[df_for_optimals['router'] == router]
        best_row_index = df_router_only[PRIMARY_METRIC].idxmin()
        best_gran = df_router_only.loc[best_row_index]['granularity']
        optimal_grans[router] = best_gran
        print(f"  Optimal granularity for '{router}': {best_gran}")
    df_fig2_filtered_list = []
    for router, gran in optimal_grans.items():
        df_slice = df_fig2_base[(df_fig2_base['router'] == router) & (df_fig2_base['granularity'] == gran)]
        df_fig2_filtered_list.append(df_slice)
    if df_fig2_filtered_list:
        df_fig2_filtered = pd.concat(df_fig2_filtered_list)
    else:
        df_fig2_filtered = pd.DataFrame()
    if not df_fig2_filtered.empty:
        fig, axes = plt.subplots(2, 2, figsize=FIG2_FIGSIZE, sharey=True)
        axes = axes.flatten()
        routers_for_plot = df_results['router'].unique().categories
        for i, router in enumerate(routers_for_plot):
            ax = axes[i]
            if router in df_fig2_filtered['router'].values:
                df_plot = df_fig2_filtered[df_fig2_filtered['router'] == router]
                sns.barplot(data=df_plot, x='model', y='MAE', ax=ax, palette='viridis')
                opt_gran = optimal_grans[router]
                ax.set_title(f'Router: {router.capitalize()} (at {opt_gran})', fontsize=FIG1_LABEL_FS)
                ax.set_xlabel(FIG2_XLABEL, fontsize=FIG2_TICK_FS)
                ax.set_ylabel(FIG2_YLABEL if i % 2 == 0 else '', fontsize=FIG2_TICK_FS)
                ax.tick_params(axis='x', rotation=45)
                ax.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
            else:
                 ax.text(0.5, 0.5, f"No data for\nrouter '{router}'", ha='center', va='center')
        fig.suptitle(FIG2_SUPTITLE, fontsize=FIG2_TITLE_FS, weight='bold')
        save_and_display_plot("fig_02_model_ablation_comparison")
    else:
        print("\n[SKIPPED] Sub-plot generation for Figure 2: Filtered DataFrame is empty.")
else:
    print("\n[SKIPPED] Figure 2: Results DataFrame is empty.")


##########################################################################################
# --- BLOCK FOR FIGURE 3: Lead Time vs. Error (Geographic Comparison) ---
##########################################################################################

#######################################################################
# CELL 7: Aesthetic Knobs for Figure 3
#######################################################################
FIG3_FIGSIZE = (16, 7)
FIG3_SUPTITLE = 'Error Growth over Prediction Horizon (RMSE)'
FIG3_XLABEL = 'Prediction Horizon (hours)'
FIG3_YLABEL = 'Root Mean Squared Error (RMSE)'
FIG3_TITLE_FS = 18
FIG3_LABEL_FS = 14
FIG3_TICK_FS = 12
FIG3_LEGEND_FS = 12
HIGH_TRAFFIC_ROUTER = 'dallas'
LOW_TRAFFIC_ROUTER = 'elpaso'

#######################################################################
# CELL 8: Code for Figure 3
#######################################################################
if not df_results.empty:
    print("\n--- Generating Figure 3: Error Growth over Horizon ---")
    
    # --- START OF FIX: Make this block self-contained ---
    optimal_gran_high = None
    optimal_gran_low = None
    # Filter for the model we use to determine optimality
    df_fig3_base = df_results[df_results['model'] == 'hybrid_gru_lstm'].copy()

    # Find optimal granularity for the high-traffic router
    df_high_base = df_fig3_base[df_fig3_base['router'] == HIGH_TRAFFIC_ROUTER]
    if not df_high_base.empty:
        # We need to find the best gran based on a consistent horizon, e.g., our FOCAL_HORIZON_HOURS
        best_row_idx = df_high_base[df_high_base['horizon_hours'] == FOCAL_HORIZON_HOURS][PRIMARY_METRIC].idxmin()
        optimal_gran_high = df_high_base.loc[best_row_idx]['granularity']
        print(f"  Using optimal granularity for {HIGH_TRAFFIC_ROUTER}: {optimal_gran_high}")
    
    # Find optimal granularity for the low-traffic router
    df_low_base = df_fig3_base[df_fig3_base['router'] == LOW_TRAFFIC_ROUTER]
    if not df_low_base.empty:
        best_row_idx = df_low_base[df_low_base['horizon_hours'] == FOCAL_HORIZON_HOURS][PRIMARY_METRIC].idxmin()
        optimal_gran_low = df_low_base.loc[best_row_idx]['granularity']
        print(f"  Using optimal granularity for {LOW_TRAFFIC_ROUTER}: {optimal_gran_low}")
    # --- END OF FIX ---
    
    fig, axes = plt.subplots(1, 2, figsize=FIG3_FIGSIZE, sharey=False)

    # --- High-Traffic Router Plot ---
    if optimal_gran_high:
        df_high = df_results[(df_results['router'] == HIGH_TRAFFIC_ROUTER) & 
                             (df_results['granularity'] == optimal_gran_high) & 
                             (df_results['model'].isin(['hybrid_gru_lstm', 'nhits']))]
        sns.lineplot(data=df_high, x='horizon_hours', y='RMSE', hue='model', marker='o', ax=axes[0])
        axes[0].set_title(f'High-Traffic: {HIGH_TRAFFIC_ROUTER.capitalize()} (at {optimal_gran_high})', fontsize=FIG3_LABEL_FS)
        axes[0].set_xlabel(FIG3_XLABEL); axes[0].set_ylabel(FIG3_YLABEL)
        axes[0].ticklabel_format(style='sci', axis='y', scilimits=(0,0))
        axes[0].set_xticks(HORIZONS_HOURS)
        axes[0].legend(title='Model', fontsize=FIG3_LEGEND_FS)
    else:
        axes[0].text(0.5, 0.5, 'No data for high-traffic router', ha='center', va='center')

    # --- Low-Traffic Router Plot ---
    if optimal_gran_low:
        df_low = df_results[(df_results['router'] == LOW_TRAFFIC_ROUTER) & 
                            (df_results['granularity'] == optimal_gran_low) &
                            (df_results['model'].isin(['hybrid_gru_lstm', 'nhits']))]
        sns.lineplot(data=df_low, x='horizon_hours', y='RMSE', hue='model', marker='o', ax=axes[1])
        axes[1].set_title(f'Low-Traffic: {LOW_TRAFFIC_ROUTER.capitalize()} (at {optimal_gran_low})', fontsize=FIG3_LABEL_FS)
        axes[1].set_xlabel(FIG3_XLABEL); axes[1].set_ylabel('')
        axes[1].ticklabel_format(style='sci', axis='y', scilimits=(0,0))
        axes[1].set_xticks(HORIZONS_HOURS)
        axes[1].legend(title='Model', fontsize=FIG3_LEGEND_FS)
    else:
        axes[1].text(0.5, 0.5, 'No data for low-traffic router', ha='center', va='center')

    fig.suptitle(FIG3_SUPTITLE, fontsize=FIG3_TITLE_FS, weight='bold')
    save_and_display_plot("fig_03_error_growth_vs_horizon")
else:
    print("\n[SKIPPED] Figure 3: Results DataFrame is empty.")


##########################################################################################
# --- BLOCK FOR FIGURE 4: Actual vs. Predicted Forecasts ---
##########################################################################################

#######################################################################
# CELL 9: Aesthetic Knobs for Figure 4
#######################################################################
FIG4_FIGSIZE = (16, 10)
FIG4_SUPTITLE = f'Qualitative Forecast Analysis for {HIGH_TRAFFIC_ROUTER.capitalize()} ({FOCAL_HORIZON_HOURS}h Horizon)'
FIG4_TITLE_FS = 18
FIG4_LABEL_FS = 14
FIG4_TICK_FS = 10
FIG4_LEGEND_FS = 10
NUM_POINTS_TO_PLOT = 200

#######################################################################
# CELL 10: Code for Figure 4
#######################################################################
if not df_results.empty:
    print("\n--- Generating Figure 4: Actual vs. Predicted Qualitative Analysis ---")
    df_fig4_base = df_results[(df_results['router'] == HIGH_TRAFFIC_ROUTER) & (df_results['horizon_hours'] == FOCAL_HORIZON_HOURS)].copy()
    df_hybrid_only = df_fig4_base[df_fig4_base['model'] == 'hybrid_gru_lstm'].copy()
    if not df_hybrid_only.empty:
        best_row_index = df_hybrid_only[PRIMARY_METRIC].idxmin()
        worst_row_index = df_hybrid_only[PRIMARY_METRIC].idxmax()
        best_gran = df_hybrid_only.loc[best_row_index]['granularity']
        worst_gran = df_hybrid_only.loc[worst_row_index]['granularity']
        print(f"  Best granularity for {HIGH_TRAFFIC_ROUTER} (hybrid model): {best_gran}")
        print(f"  Worst granularity for {HIGH_TRAFFIC_ROUTER} (hybrid model): {worst_gran}")
        scenarios = {
            'Top-Left: Hybrid Model (Optimal Gran.)': {'model': 'hybrid_gru_lstm', 'gran': best_gran},
            'Top-Right: Hybrid Model (Sub-Optimal Gran.)': {'model': 'hybrid_gru_lstm', 'gran': worst_gran},
            'Bottom-Left: N-HiTS Model (Optimal Gran.)': {'model': 'nhits', 'gran': best_gran},
            'Bottom-Right: N-HiTS Model (Sub-Optimal Gran.)': {'model': 'nhits', 'gran': worst_gran},
        }
        fig, axes = plt.subplots(2, 2, figsize=FIG4_FIGSIZE, sharex=True, sharey=True)
        axes = axes.flatten()
        for i, (title, params) in enumerate(scenarios.items()):
            ax = axes[i]
            try:
                filename = f"{HIGH_TRAFFIC_ROUTER}_{params['gran']}_{params['model']}.parquet"
                df_preds = pd.read_parquet(PREDICTIONS_DIR / filename)
                h_steps = hours_to_steps(FOCAL_HORIZON_HOURS, params['gran'])
                true_col = f'y_true_step_{h_steps}'
                pred_col = f'y_pred_step_{h_steps}'
                if true_col in df_preds.columns and pred_col in df_preds.columns:
                    plot_data = df_preds.head(NUM_POINTS_TO_PLOT)
                    ax.plot(plot_data['timestamp'], plot_data[true_col], label='Actual', color='blue', alpha=0.8)
                    ax.plot(plot_data['timestamp'], plot_data[pred_col], label='Predicted', color='red', linestyle='--')
                    ax.set_title(title, fontsize=FIG4_LABEL_FS)
                    ax.tick_params(axis='x', rotation=45, labelsize=FIG4_TICK_FS)
                    ax.legend(fontsize=FIG4_LEGEND_FS)
                    ax.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
                else:
                    ax.text(0.5, 0.5, f"Horizon {FOCAL_HORIZON_HOURS}h data\nnot found in file", ha='center', va='center')
            except FileNotFoundError:
                ax.text(0.5, 0.5, f"Prediction file not found for\n{params['model']} at {params['gran']}", ha='center', va='center', wrap=True)
            except Exception as e:
                ax.text(0.5, 0.5, f"Error plotting: {e}", ha='center', va='center', wrap=True)
        fig.suptitle(FIG4_SUPTITLE, fontsize=FIG4_TITLE_FS, weight='bold')
        save_and_display_plot("fig_04_qualitative_forecasts")
    else:
        print(f"\n[SKIPPED] Sub-plot generation for Figure 4: No data found for model 'hybrid_gru_lstm' to determine optimal granularities.")
else:
    print("\n[SKIPPED] Figure 4: Results DataFrame is empty.")


##########################################################################################
# --- BLOCK FOR TABLE 1: Optimal Granularity Summary ---
##########################################################################################

#######################################################################
# CELL 11: Code for Table 1
#######################################################################
if not df_results.empty:
    print("\n--- Generating Table 1: Optimal Granularity and Performance Summary ---")
    df_table1 = df_results[(df_results['horizon_hours'] == FOCAL_HORIZON_HOURS) & (df_results['model'] == 'hybrid_gru_lstm')]
    summary_list = []
    for router in df_table1['router'].unique():
        df_router = df_table1[df_table1['router'] == router]
        best_row = df_router.loc[df_router[PRIMARY_METRIC].idxmin()]
        worst_row = df_router.loc[df_router[PRIMARY_METRIC].idxmax()]
        gain = ((worst_row[PRIMARY_METRIC] - best_row[PRIMARY_METRIC]) / worst_row[PRIMARY_METRIC]) * 100
        summary_list.append({
            'Router': router.capitalize(),
            'Optimal Granularity': best_row['granularity'],
            f'Best {PRIMARY_METRIC} (%)': f"{best_row[PRIMARY_METRIC]:.2f}",
            f'Worst {PRIMARY_METRIC} (%)': f"{worst_row[PRIMARY_METRIC]:.2f}",
            'Performance Gain (%)': f"{gain:.1f}"
        })
    table1_df = pd.DataFrame(summary_list)
    display(table1_df)
else:
    print("\n[SKIPPED] Table 1: Results DataFrame is empty.")

##########################################################################################
# --- BLOCK FOR TABLE 2: Comprehensive Model Performance ---
##########################################################################################

#######################################################################
# CELL 12: Code for Table 2
#######################################################################
if not df_results.empty:
    print("\n--- Generating Table 2: Comprehensive Model Performance (Median Values) ---")
    median_performance = df_results.groupby('model')[['MAE', 'RMSE', 'sMAPE', 'R2', 'MSLE', 'P99_abs_error', 'training_time_sec']].median()
    pd.options.display.float_format = '{:.2f}'.format
    display(median_performance.sort_values(by='MAE'))
else:
    print("\n[SKIPPED] Table 2: Results DataFrame is empty.")


#######################################################################
# CELL 13: COMPLETION
#######################################################################
print("\n--- Notebook 04_Results_and_Visualization.ipynb Complete ---")
print(f"All final figures have been saved to the '{FINAL_FIGURES_DIR}' directory.")
```
